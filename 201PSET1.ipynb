### **Question 1a**


Let's take the polynomial

$$p_A(x) = a_1x^9-a_2x^3+a_3=0$$

To nondimensionalize this equation, we take $x=(\frac{a_3}{a_2})^{\frac{1}{3}}y$ 

Substituting into the original polynomial equation, this gives us $$a_1(\frac{a_3}{a_2})^{3}y^9-a_2\frac{a_3}{a_2}y^3+a_3=0$$

Simplifying, we have $$\frac{a_1 a_3^{3}}{a_2^{3}}y^9-a_3y^3+a_3=0$$

Dividing through by $a_3$, we have $$\frac{a_1 a_3^{2}}{a_2^{3}}y^9-y^3+1=0$$

Now we take the single dimensionless parameter $\epsilon = \frac{a_1 a_3^{2}}{a_2^{3}}$, which gives us

$$q_{A}(y) = \epsilon y^7 - y^3 +1 =0$$

### **Question 1b**

We apply the method of dominant balances. Let us take $$A = \epsilon y^9, B = y^3, C = 1$$

Then the equation reduces to $$A-B+C=0$$.

*Case 1*

Let us take $|A| \sim |B| >> |C|$

This gives $\epsilon y^9 = y^3$

We know this is true for $y=0$ and $y=(\frac{1}{\epsilon})^\frac{1}{6}$

Let us check consistency of the balance in either case. For $y=0$, we have $ \epsilon y^9 = y^3 = 0$, but this is less than 1 so the balance is inconsistent. We disregard $y=0$.

Next, for $y=(\frac{1}{\epsilon})^\frac{1}{6}$, i.e. $(\frac{1}{\epsilon^\frac{1}{6}},\frac{-1}{\epsilon^\frac{1}{6}},\frac{-1-i\sqrt{3}}{2\epsilon^\frac{1}{6}}\frac{-1+i\sqrt{3}}{2\epsilon^\frac{1}{6}}, \frac{1-i\sqrt{3}}{2\epsilon^\frac{1}{6}}, \frac{1+i\sqrt{3}}{2\epsilon^\frac{1}{6}})$,

$$|A| = |\frac{1}{\epsilon^\frac{2}{3}}|$$

$$|B| = |\frac{1}{\epsilon^\frac{2}{3}}|$$

$$|C| = 1$$

Therefore, this balance is consistent only for $\epsilon \to 0$, **not** for $\epsilon \to \infty.$

*Case 2*

Let us take $|A| \sim |C| >> |B|$

This gives $\epsilon y^9 = 1$, hence $y = (\frac{1}{\epsilon})^\frac{1}{9}$

The ninth roots of 1 are $$e^\frac{i \pi}{9}, e^\frac{3i \pi}{9},e^\frac{5i \pi}{9},e^\frac{7i \pi}{9},-1, e^\frac{-i \pi}{9},e^\frac{-3i \pi}{9},e^\frac{-5i \pi}{9}, e^\frac{-7i \pi}{9}$$

So this gives us 7 complex and 1 real root

Now we check for consistency of this balance:

$$|A| = |\epsilon \times \frac{1}{\epsilon^{\frac{1}{9} \times 9}}| = 1$$

$$|B| = |\frac{1}{\epsilon^\frac{3}{9}}|$$

$$|C| = 1$$

Therefore the balance $|A| \sim |C| >> |B|$ holds only for $\epsilon \to \infty$, **not** $\epsilon \to 0$.

*Case 3*

Take $|B| \sim |C| >> |A|$

This gives $y^3 = 1$, so $y = 1^\frac{1}{3}$

Hence y are the cube roots of unity $y = (e^\frac{2 \pi i}{3}, e^\frac{-2 \pi i}{3}, 1)$

Now we check for consistency of this balance:

$$|A| = |\epsilon|$$

$$|B| = 1$$

$$|C| = 1$$

Hence this balance holds only for $\epsilon \to 0$.

Hence we have found our solutions:

$y = \begin{cases}
 \frac{e^\frac{i \pi}{9}}{\epsilon^\frac{1}{9}}, \frac{e^\frac{3i \pi}{9}}{{\epsilon^\frac{1}{9}}},\frac{e^\frac{5i \pi}{9}}{{\epsilon^\frac{1}{9}}},\frac{e^\frac{7i \pi}{9}}{{\epsilon^\frac{1}{9}}},\frac{-1}{{\epsilon^\frac{1}{9}}}, \frac{e^\frac{-i \pi}{9}}{{\epsilon^\frac{1}{9}}},\frac{e^\frac{-3i \pi}{9}}{{\epsilon^\frac{1}{9}}},\frac{e^\frac{-5i \pi}{9}}{{\epsilon^\frac{1}{9}}},\frac{e^\frac{-7i \pi}{9}}{{\epsilon^\frac{1}{9}}} & ,\epsilon \to \infty \\
 e^\frac{2 \pi i}{3}, e^\frac{-2 \pi i}{3}, 1,\frac{1}{\epsilon^\frac{1}{6}},\frac{-1}{\epsilon^\frac{1}{6}},\frac{-1-i\sqrt{3}}{2\epsilon^\frac{1}{6}}\frac{-1+i\sqrt{3}}{2\epsilon^\frac{1}{6}}, \frac{1-i\sqrt{3}}{2\epsilon^\frac{1}{6}}, \frac{1+i\sqrt{3}}{2\epsilon^\frac{1}{6}} & ,\epsilon \to 0
\end{cases}$

This means that for $\epsilon \to \infty$, we have eight complex roots and one real root, and for $\epsilon \to 0$, we have six complex roots and three real roots.
